## Project Summary - 

This project aims to perform an Exploratory Data Analysis (EDA) on Amazon Prime Video content to uncover meaningful insights about content trends, user preferences, and platform strategy. The analysis focuses on a dataset that includes all shows available on Amazon Prime Video in the United States. It comprises two CSV files: titles.csv and credits.csv, containing a mix of categorical and numerical attributes.

The titles.csv file includes over 9,000 unique titles available on Amazon Prime. It has 15 columns that provide detailed metadata about each title, such as id, title, show_type (movie or TV show), description, release_year, age_certification, runtime, genres, production_countries, seasons, imdb_id, imdb_score, imdb_votes, tmdb_popularity, and tmdb_score. These features offer a comprehensive view of the content's nature, release timeline, audience rating, and popularity on both IMDb and TMDb platforms.

The second file, credits.csv, contains more than 124,000 records related to the cast and crew involved in these titles. This dataset includes five columns: person_id, id (which links back to the title), name, character_name, and role (such as actor or director).

The first step of the analysis involved understanding the dataset by checking the data types of each column, identifying missing and duplicate values, and summarizing the numerical attributes through descriptive statistics. Data cleaning and preprocessing were essential to ensure accuracy and consistency across both datasets.

The second step is to apply various EDA techniques, including data aggregation, grouping, filtering, and visualizations. These techniques help uncover hidden patterns and trends, such as the dominance of specific genres like drama or action, the impact of production countries. Visual tools like bar charts, pie charts, histograms and heatmaps are used to present insights effectively.

Overall, this project provides a clear view of how Amazon Prime Video's content is structured and how it aligns with viewer behavior and preferences. The findings can help inform business decisions related to content acquisition, recommendation systems, and audience targeting.

## Data Wrangling

```python
# remove duplicates
merged_df = merged_df.drop_duplicates()
```

```python
merged_df.info()
```

```python
# drop the columns seasons
merged_df = merged_df.drop('seasons', axis=1)
```

```python
merged_df.columns
```

```python
# fill missing value for age_certification with 'unknown'
merged_df['age_certification'] = merged_df['age_certification'].fillna('Unknown')
```

```python
merged_df['age_certification']
```

```python
# fill missing values of character with mode
character_mode = merged_df['character'].mode()[0]
merged_df['character'] = merged_df['character'].fillna(character_mode)
```

```python
merged_df['character']
```

```python
# remove the rows of description, imdb_id, imdb_score, imdb_votes, tmdb_popularity, tmdb_score columns which have null/missing values.
merged_df.dropna(subset=['description', 'imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score'], inplace=True)
```

```python
merged_df.info()
```

- Remove duplicate records.
- Drop seanons columns, 93.44% values are missing in this column.
- Fill the missing value in age-certification column with 'unknown'.
- Fill the missing value in charater column with mode.
- Remove the rows of description, imdb_id, imdb_score, imdb_votes, tmdb_popularity, tmdb_score columns which have null/missing values.
- Insights: clean, non-duplicated dataset with no critical missing values.

## Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

```python
# Content Category
merged_df['type'].value_counts().plot.bar()
plt.title('Content Category Distribution')
plt.show()
```

1. Why did you pick this specific chart?
   Bar chart visulaize very clearly that movie content dominate the plateform.

2. What is/are the insight(s) found from the chart?
   Strong user preference or higher content supply for movies, indicating that the platform may be more focused on entertainment.

3. Will the gained insights help creating a positive business impact?
   Yes, insights support better resource allocation, smarter content planning all of which contribute to improved user experience and long-term business growth.
   

```python
# Content Genres
genres_series = merged_df['genres'].dropna()
genres_series = genres_series.str.replace(r"[\[\]']", "", regex=True)
genres_exploded = genres_series.str.split(',').explode()
genres_exploded = genres_exploded.str.strip().str.title()
genres_counts = genres_exploded.value_counts()

plt.figure(figsize=(12,6))
genres_counts.head(10).plot(kind='barh')
plt.title('Top 10 Genres of Amazon Prime')
plt.show()

   
          
```

1. Why did you pick this specific chart? A horizontal bar chart was selected because it is effective for displaying text-heavy categories like genre names. 

2. What is/are the insight(s) found from the chart? The chart shows the top 10 most frequent genres on Amazon Prime. It reveals what type of content is most prevalent—such as Drama, Comedy, or Action—giving insight into the platform’s content focus.

3. Will the Gained Insights Help Creating a Positive Business Impact? genre analysis informs smarter content and marketing strategies, helping improve viewer satisfaction and platform performance.
   

```python
# Content Distribution in different regions
country_series = merged_df['production_countries'].dropna()
country_series = country_series.str.replace(r"[\[\]']", "", regex=True)
all_countries = country_series.str.split(',').explode()
all_countries = all_countries.str.strip()


plt.figure(figsize=(10,6))
all_countries.value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries with Most Content Availability')
plt.ylabel('Number of Titles')
plt.show()
```

Why did you pick this specific chart? A bar chart was chosen because it clearly shows the top 10 countries contributing the most content on the platform.

What is/are the insight(s) found from the chart? The chart reveals which countries are the largest content contributors on the platform. It highlights the geographic diversity of content, showing whether the platform sources mostly from a few countries or has broad international representation.

Will the gained insights help creating a positive business impact? Yes, insights support understanding country-level content distribution helps improve strategic decisions around content sourcing, audience targeting, and global growth.

```python
# Top rated title
top_rated = merged_df.dropna(subset=['imdb_score'])  
top_rated = top_rated.sort_values(by='imdb_score', ascending=False).head(20)
plt.figure(figsize=(12,6))
plt.barh(top_rated['title'], top_rated['imdb_score'], color='mediumseagreen')
plt.xlabel('IMDb Rating')
plt.title('Top 20 Highest Rated Shows on Amazon Prime')
plt.gca().invert_yaxis()  
plt.grid(axis='x')
plt.show()
```

Why did you pick this specific chart? A horizontal bar chart was chosen because it effectively displays the top 20 highest rated shows in a clear, easy-to-read format.

What is/are the insight(s) found from the chart? The chart highlights which shows are most highly rated by viewers. It helps identify top-performing content that could be key drivers for audience engagement.

Will the gained insights help creating a positive business impact? Yes, insights support showcasing and focusing on highly rated content can boost the platform’s reputation, viewer loyalty, and revenue growth.

```python
# trends over time

releases_per_year = merged_df.groupby('release_year').size()

plt.figure(figsize=(10,5))
releases_per_year.plot(kind='line', marker='o')
plt.title('Content Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

Why did you pick this specific chart? A line chart was chosen because it clearly shows how the number of content releases changes over time. Line chart is perfect for visualizing trends, growth, or declines year by year, helping to easily spot patterns or shifts in content production.

What is/are the insight(s) found from the chart? The chart shows whether content releases are increasing, decreasing, or stable over the years. This helps understand the platform’s content growth and investment trends over time.

Will the gained insights help creating a positive business impact? Yes, insights support to tracking content releases over time enables better planning and strategy, leading to stronger business performance and customer satisfaction.

```python
# IMdb rating vs popularity
plt.figure(figsize=(12,6))
sns.scatterplot(data=merged_df, x='imdb_score', y='tmdb_popularity')
plt.title('IMDB Rating vs TMDB Popularity')
plt.xlabel('IMDB Rating')
plt.ylabel('TMDB Popularity')
plt.grid(True)
plt.show()
```

Why did you pick this specific chart? A scatter plot was chosen because it’s perfect for showing the relationship between two continuous variables — here, IMDb ratings and TMDB popularity.

What is/are the insight(s) found from the chart? The scatter plot reveals if there is any correlation between ratings and popularity. Outliers can be identified — titles that are either very popular but with low ratings or vice versa.

Will the gained insights help creating a positive business impact? Yes, this analysis enables smarter decisions that improve viewer engagement and satisfaction, positively impacting the business.

```python
# content type vs release year
content_release_year = merged_df['release_year'].value_counts().head(10).index
merged_df[merged_df['release_year'].isin(content_release_year)].groupby(['release_year','type']).size().unstack().plot(kind='bar', stacked=True)
plt.show()
```

Why did you pick this specific chart? A stacked bar chart was chosen because it clearly shows the composition of content types (e.g., Movies and TV Shows) released in the top 10 most active years. Understand trends and shifts in content strategy over time.

What is/are the insight(s) found from the chart? identify which years had the highest volume of content released. It shows how the platform’s focus on Movies vs TV Shows changed during the busiest years.

Will the gained insights help creating a positive business impact? Yes, understanding content release trends helps the platform optimize its strategy, attract more viewers, and improve overall business performance.

```python
# top 10 type and role
plt.figure(figsize=(10, 6))
sns.countplot(x='role', hue='type', data=merged_df)
plt.title('Count of content type and role')
plt.xlabel('Ratings')
plt.xticks(rotation = 45)
plt.show()
```

Why did you pick this specific chart? A countplot was chosen because it’s great for showing the number of occurrences of different categories—in this case, the roles (like actor, director, etc.) and how they are distributed across content types (e.g., Movie, TV Show). 

What is/are the insight(s) found from the chart? You can see which roles appear most frequently on the platform overall. It highlights any role imbalance or dominance in specific content types, which may reveal production focus or audience preferences.

Will the gained insights help creating a positive business impact? Yes, this analysis supports smarter resource allocation and improves content quality, benefiting the platform’s growth and viewer satisfaction.

```python
# top 10 name with role
df_clean = merged_df.dropna(subset=['name', 'role'])
name_role_counts = df_clean.groupby(['name', 'role']).size().reset_index(name='count')
top_10_names = name_role_counts.groupby('name')['count'].sum().sort_values(ascending=False).head(10).index
top_10_df = name_role_counts[name_role_counts['name'].isin(top_10_names)]

plt.figure(figsize=(12,6))
sns.barplot(data=top_10_df, x='name', y='count', hue='role', palette='Set2')
plt.title('Top 10 Names by Role')
plt.xlabel('character')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.legend(title='Role')
plt.tight_layout()
plt.show()
```

Why did you pick this specific chart? Bar charts are excellent for comparing discrete categories like names and roles.

What is/are the insight(s) found from the chart? The chart shows that some people, like actors or directors, are involved in many titles on the platform. This means they are either very popular or work a lot.

Will the gained insights help creating a positive business impact? Yes, this information helps the platform create better content and keep viewers happy, which is good for business.

```python
# content type release over the year
content_by_year = merged_df.groupby(['release_year', 'type']).size().unstack(fill_value=0)
content_by_year.tail(10).plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Content Type Added Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

Why did you pick this specific chart? It clearly shows increases or decreases in content production over time. The marker='o' makes each data point stand out for easier year-by-year analysis.

What is/are the insight(s) found from the chart? A significant increase in movie content was observed from 2016 to 2021, indicating a possible strategic push. The content release trend might peak around a certain year and then decline, possibly reflecting market saturation or shifting strategies.



```python
# Top 10 title by rating and type
top_titles = merged_df.dropna(subset=['imdb_score'])
top_titles = top_titles.drop_duplicates(subset=['title'])
top_titles = top_titles.sort_values(by='imdb_score', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_titles, x='imdb_score', y='title', hue='type', dodge=False, palette='Set2')
plt.title('Top 10 Highest Rated Titles by Type')
plt.xlabel('IMDb Rating')
plt.ylabel('Title')
plt.legend(title='Content Type')
plt.tight_layout()
plt.show()
```

1. Why did you pick the specific chart?
   A horizontal bar chart is ideal for comparing IMDb ratings across a small number of categories with long names like movie or show titles. The use of hue='type' helps quickly distinguish between Movies and TV Shows within the top 10.

2. What is/are insight(s) found from the chart?
   The chart reveals the highest-rated content on the platform, based on IMDb scores.It may highlight specific titles that stand out with exceptionally high ratings.
  

```python
# Heatmap: Year vs Content type
heatmap_data = pd.pivot_table(merged_df, index='type', columns='release_year', aggfunc='size', fill_value=0)
plt.figure(figsize=(15, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Netflix Content Type Added per Year')
plt.xlabel('Release Year')
plt.ylabel('Content Type')
plt.show()    
```

1. Why did you pick the specific chart?
    visually compare content trends over time across different content types in one compact visual. The use of color intensity makes it easy to spot  peaks and declines.

2. What is/are insight(s) found from the chart?
   Movie content has been released in higher volume than TV shows every year. A significant increase in movie content was observed from 2016 to 2021.
